```python
import torch
import cv2 # video frame extraction
from pathlib import Path
import numpy as np
import os
```

# DFDC Data Preprocess

## Extract frame image from dataset

```python
def extractAllFiles(videoFolderPath: str, imageFolderPath: str):
    
    p = Path(videoFolderPath)
    for file in p.iterdir():
        
        fileName = file.name[:-4]
        if file.name[-3:0] != 'mp4': continue # skip .json file
        videoObject = cv2.VideoCapture(str(file))
        
        imagePath = imageFolderPath + fileName
        if not os.path.exists(imagePath):
            os.mkdir(imagePath)
            print("Directory " , imagePath ,  " Created ")
        else:
            print("Directory " , imagePath ,  " already exists")
            
        frameCount = 0
        success = True
        while success:
            success, frame = videoObject.read()
            if not success:break
            outputImageName = f"{imagePath}\\frame_{frameCount}.jpg"
            cv2.imwrite(outputImageName, frame) #"frame%d.jpg" % frameCount
            frameCount += 1
        
        print(f"File {file.name} done extraction")
    print(f"All files extraction complete")
```

```python
cwd = os.getcwd()
videoFolderPath = cwd+"\\dfdc_train_part_00\\dfdc_train_part_0\\"
imageFolderPath = cwd + "\\frameExtraction\\"
extractAllFiles(videoFolderPath, imageFolderPath)
```
